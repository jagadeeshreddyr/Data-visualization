```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = [8, 6]
plt.style.use('tableau-colorblind10')
```

# Bar Graph

```python
d = {'2011':[100000, 34651, 24500, 0], 
     '2013':[100000, 34561, 24500, 0],
     '2017':[100000, 38000, 26500, 1000]}
df = pd.DataFrame(data=d,columns=['2011', '2013', '2017'], index = ['Protected Forest', 'Total Forest Area', 'Reserved Forests', 
                                                'Un classed Forest'])
df.plot.bar()
plt.xticks(rotation=0)
plt.title('Total Recored Forest (Uttarakhand)')
plt.ylabel('KM^2')
plt.xlabel('Types of Forest')
plt.savefig('Visual/Bar.png')
```

## Line Graph

```python
d = {'Year':[2001, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017,2019],
    '% Coverage':[43.53, 45.55, 45.53, 45.58, 45.57, 45.58, 45.58, 45.39, 46.20,45.00]}
df2 = pd.DataFrame(data =d)

line_color = 'blue'  # Change to the desired color
line_width = 2.0  # Change to the desired line width
marker_style = 'o'  # Change to the desired marker style
marker_size = 8  # Change to the desired marker size

# Plotting
plt.plot(df2['Year'], df2['% Coverage'], color=line_color, linewidth=line_width, marker=marker_style, markersize=marker_size)

# Customize other plot elements as needed
odd_years = df2['Year'][df2['Year'] % 2 != 0].tolist()
plt.xlabel('Year',fontsize = 12)
plt.ylabel('% Coverage',fontsize = 12)
plt.title('Area of Forest Coverage Over Years')
plt.ylim(40,50)
plt.xticks(odd_years, fontsize=12)  # Adjust fontsize as needed for tick labels
plt.yticks(fontsize=12)  

# Show the plot
# plt.show()
plt.savefig('Visual/Line.png')
```

# Horizontal Bar Graph

```python

d = {'Survival% 2016-2017':[12180, 2700, 1500, 3500,1800,1700,300,1980],
 'Survival% 2017-2018':[12200, 2050, 1150,3000,1980,2050,500,1900]}

df8 = pd.DataFrame(d, index=['Total','West Region', 'South Kumau Region', 'North Kumau Region', 'Shiwalik Region',
                            'Bhagirathi Region', 'Yamuna Region', 'Garhwai Region'])
plt.figure(figsize=(20, 6), dpi=80)
plt.style.use('tableau-colorblind10')
plt.xticks([0,100])
df8.plot.barh()
# plt.plot.barh(df3[['']])
plt.xlabel('Area (ha)', fontsize = 12)

plt.yticks(rotation=0)
plt.ylabel('Forest Divisions', fontsize = 12)
plt.title('Plantation raised in Area (ha)',fontsize = 14)
plt.legend(['2016-2017', '2017-2018'], bbox_to_anchor=(0.9, 1.05), loc='upper left')
plt.savefig('Visual/horizontal_Bar.png')
```

## Pie Graph
